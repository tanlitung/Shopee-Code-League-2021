```python
import numpy as np
import pandas as pd
import json
from tqdm import tqdm
```

```python
df = pd.read_json('Dataset for Multi-Channel Contacts Problem/contacts.json')
```

```python
df.head()
```

```python
class User():
    def __init__(self):
        self.email = []
        self.phone = []
        self.order = []
        self.contacts = 0
        self.ticket_id = []
```

```python
email_dict = {}
order_dict = {}
phone_dict = {}
users = []
for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    user = None
    temp_email = 0
    temp_phone = 0
    temp_order = 0
    if row['Email'] != None:
        temp_email = email_dict.get(row['Email'])
    if row['Phone'] != None:
        temp_phone = phone_dict.get(row['Phone'])
    if row['OrderId'] != None:
        temp_order = order_dict.get(row['OrderId'])
        
    if (temp_email == None) or (temp_phone == None) or (temp_order == None):
        user = User()
        user.email.append(row['Email'])
        user.phone.append(row['Phone'])
        user.order.append(row['OrderId'])
        user.contacts += int(row['Contacts'])
        user.ticket_id.append(row['Id'])
        users.append(user)
        
        if (temp_email == None):
            email_dict[row['Email']] = user
            
        if (temp_phone == None):
            email_dict[row['Phone']] = user
            
        if (temp_order == None):
            email_dict[row['OrderId']] = user

    if (temp_email != None) and (temp_phone == None) and (temp_order == None):
        temp_email.email.append(row['Email'])
        temp_email.phone.append(row['Phone'])
        temp_email.order.append(row['OrderId'])
        temp_email.contacts += int(row['Contacts'])
        temp_email.ticket_id.append(row['Id'])

    elif (temp_email == None) and (temp_phone != None) and (temp_order == None):
        temp_phone.email.append(row['Email'])
        temp_phone.phone.append(row['Phone'])
        temp_phone.order.append(row['OrderId'])
        temp_phone.contacts += int(row['Contacts'])
        temp_phone.ticket_id.append(row['Id'])

    elif (temp_email == None) and (temp_phone == None) and (temp_order != None):
        temp_order.email.append(row['Email'])
        temp_order.phone.append(row['Phone'])
        temp_order.order.append(row['OrderId'])
        temp_order.contacts += int(row['Contacts'])
        temp_order.ticket_id.append(row['Id'])

    elif (temp_email != None) and (temp_phone != None) and (temp_order == None):
        user = User()
        user.email.append(list(set(temp_email.email + temp_phone.email)))
        user.phone.append(list(set(temp_email.phone + temp_phone.phone)))
        user.order.append(list(set(temp_email.order + temp_phone.order))).append(row['OrderId'])
        user.contacts += (int(temp_email.contacts) + int(temp_phone.contacts) + int(row['Contacts']))
        user.ticket_id.append(list(set((temp_email.ticket_id + temp_phone.ticket_id).append(row['Id']))))
        
        users = filter(lambda user: row['Email'] not in user.email, users)
        users = filter(lambda user: row['Phone'] not in user.phone, users)
        users.append(user)
        
        email_dict[row['Email']] = user
        phone_dict[row['Phone']] = user
        
        
    elif (temp_email == None) and (temp_phone != None) and (temp_order != None):
        user = User()
        user.email.append(list(set(temp_order.email + temp_phone.email))).append(row['Email'])
        user.phone.append(list(set(temp_order.phone + temp_phone.phone)))
        user.order.append(list(set(temp_order.order + temp_phone.order)))
        user.contacts += (int(temp_order.contacts) + int(temp_phone.contacts) + int(row['Contacts']))
        user.ticket_id.append(list(set((temp_order.ticket_id + temp_phone.ticket_id).append(row['Id']))))
        
        users = filter(lambda user: row['OrderId'] not in user.order, users)
        
        users.append(user)
        
        order_dict[row['OrderId']] = user
        phone_dict[row['Phone']] = user
        
    elif (temp_email != None) and (temp_phone == None) and (temp_order != None):
        user = User()
        user.email.append(list(set(temp_email.email + temp_order.email)))
        user.phone.append(list(set(temp_email.phone + temp_order.phone))).append(row['OrderId'])
        user.order.append(list(set(temp_email.order + temp_order.order)))
        user.contacts += (int(temp_email.contacts) + int(temp_order.contacts) + int(row['Contacts']))
        user.ticket_id.append(list(set((temp_email.ticket_id + temp_order.ticket_id).append(row['Id']))))
                       
        users = filter(lambda user: row['Email'] not in user.email, users)
        users = filter(lambda user: row['OrderId'] not in user.order, users)
        users.append(user)
        
        email_dict[row['Email']] = user
        order_dict[row['OrderId']] = user
        
    elif (temp_email != None) and (temp_phone != None) and (temp_order != None):
        user = User()
        user.email.append(list(set(temp_email.email + temp_order.email + temp_phone.email)))
        user.phone.append(list(set(temp_email.phone + temp_order.phone + temp_phone.phone)))
        user.order.append(list(set(temp_email.order + temp_order.order + temp_phone.order)))
        user.contacts += (int(temp_email.contacts) + int(temp_order.contacts) + int(temp_phone.contacts) + int(row['Contacts']))
        user.ticket_id.append(list(set((temp_email.ticket_id + temp_phone.ticket_id + temp_order.ticket_id).append(row['Id']))))
        
        users = filter(lambda user: row['Email'] not in user.email, users)
        users = filter(lambda user: row['OrderId'] not in user.order, users)
        users = filter(lambda user: row['Phone'] not in user.phone, users)
        users.append(user)
        
        email_dict[row['Email']] = user
        phone_dict[row['Phone']] = user
        order_dict[row['OrderId']] = user
```

```python
csv_dict = {}
for row, user in tqdm(enumerate(users)):
    ticket_ids = sorted(user.ticket_id)
    list_string = list(map(str, ticket_ids))
    str1 = '-'.join(list_string)
    csv_dict[row] = str1 + ', ' + str(user.contacts)
```

```python
df1 = pd.DataFrame.from_dict(csv_dict, orient='index', columns = ['ticket_trace/contact'])
df1.index.name = 'ticket_id'
df1.to_csv("output.csv")
```
